```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset
df = pd.read_csv('C:\\Users\\suyog\\Downloads\\disney_plus_titles.csv')
```

```python
df_cleaned = df.fillna('Unknown')
```

```python
# 1. What is the distribution of content types (Movies vs TV Shows)?
plt.figure(figsize=(10, 6))
sns.countplot(data=df_cleaned, x='type')
plt.title('Count of Movies vs TV Shows')
plt.xlabel('Type')
plt.ylabel('Count')
plt.show()
```

```python
# 2. Which country has produced the most content?
country_counts = df_cleaned['country'].value_counts()
print("Country with most content:\n", country_counts)
```

```python
# 3. What is the distribution of release years?
plt.figure(figsize=(10, 6))
sns.histplot(df_cleaned['release_year'], kde=True)
plt.title('Distribution of Release Year')
plt.xlabel('Release Year')
plt.ylabel('Frequency')
plt.show()
```

```python
# 4. What are the top 5 genres listed in the dataset?
genres = df_cleaned['listed_in'].str.split(', ').explode()
top_genres = genres.value_counts().head(5)
print("Top 5 genres:\n", top_genres)
```

```python
# 5. Which directors have directed the most titles?
top_directors = df_cleaned['director'].str.split(', ').explode().value_counts().head(5)
print("Top 5 directors:\n", top_directors)
```

```python
#graph of the correlation matrix
plt.figure(figsize=(12, 8))
correlation_matrix = df_cleaned.corr(numeric_only=True)
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```
